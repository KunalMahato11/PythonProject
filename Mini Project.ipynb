2. Plot a bar graph no. of matches won by each team in the year 2011

```python
import matplotlib.pyplot as plt
import numpy as np
x = pd.DataFrame(df[df['Season_Year'] == 2011])
z = x["Team1"].unique()
y = pd.DataFrame(x['match_winner'].value_counts())
a = []
for i in range(len(y)):
    a.append(y['match_winner'].values[i])
index = np.arange(len(a))
width = plt.barh(z,a,color='blue')
plt.show()
```
